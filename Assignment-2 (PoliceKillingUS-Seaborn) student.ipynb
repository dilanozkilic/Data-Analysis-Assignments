## 2: Plot 12 most dangerous cities

```python
cities = kill.city.value_counts()[:12]
```

```python
cities
```

```python
# with seaborn barplot

plt.figure(figsize=(8,6))
sns.barplot(x=cities.index, y=cities.values)
plt.title("Most Dangerous Cities")
plt.xticks(rotation = 45);
```

```python
# with seaborn countplot

plt.figure(figsize=(8,6))
sns.countplot(x="city",order=kill.city.value_counts()[:12].index, data=kill)
plt.title("Most Dangerous Cities")
plt.xticks(rotation = 45);
```

## 3: Plot 20 most dangerous states

```python
states = kill.state.value_counts()[:20]
states
```

```python
# with seaborn barplot

plt.figure(figsize=(8,6))
sns.barplot(x=states.index, y=states.values)
plt.title("Most Dangerous States")
plt.xticks(rotation = 45);
```

```python
# with seaborn countplot

plt.figure(figsize=(8,6))
sns.countplot(x="state",order=kill.state.value_counts()[:20].index, data=kill)
plt.title("Most Dangerous States")
plt.xticks(rotation = 45);
```
